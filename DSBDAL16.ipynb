```python
import pandas as pd
```

```python
import numpy as np
```

```python
import string
```

```python
import re
```

```python
import nltk
```

```python
df = pd.read_csv('Alexa-Dataset - Alexa-Dataset.csv')
df
```

```python
alexa_text = df.drop(['rating','feedback','date','variation'],axis=1)
alexa_text
```

```python
alexa_text_lower = alexa_text.apply(lambda x:x.astype(str).str.lower())
```

```python
df['lower_text'] = alexa_text_lower
```

```python
def remove_punctuation(text):
    punctuationfree = "".join([word for word in text if word not in string.punctuation])
    return punctuationfree
df['rem_punc'] = df['lower_text'].apply(lambda x:remove_punctuation(x))
df
```

```python
def tokenization(text):
    token = re.split('W+',text)
    return token
df['tokenized_string'] = df['lower_text'].apply(lambda x:tokenization(x))
df
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
",".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
df['rem_stopword'] = df['lower_text'].apply(lambda x:remove_stopwords(x))
df
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def stemming(text):
    return " ".join([stemmer.stem(word) for word in text.split()])
df['stemmend_text'] = df['lower_text'].apply(lambda x:stemming(x))
df
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatization(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
df['lemmatized_text'] = df['lower_text'].apply(lambda x:lemmatization(x))
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow = CountVectorizer()
```

```python
X_bow = bow.fit_transform(df['lower_text'])
print(X_bow)
```

```python
X_tfidf = tfidf.fit_transform(df['lower_text'])
print(X_tfidf)
```
