```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
df.drop('deck',axis=1)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data=df)
```

```python
sns.countplot(data=df,x='sex',hue='survived')
```

```python
sns.histplot(data=df,x='fare',kde=True,bins=40)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('No. of passengers')
plt.show()
```

```python
sns.histplot(data=df, x='sex', hue='survived', multiple='stack', kde=True)
plt.title('Age Distrib')
```

```python
sns.boxplot(data=df, x='sex', y='age', hue='survived')
plt.title('Age Distribution by Sex and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived')
plt.show()
```
