```python
def detect_outliers(data):
    outliers = []
    data = sorted(data)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1

    print("Q1:", Q1)
    print("Q3:", Q3)
    print("IQR:", IQR)

    lower_bnd = Q1 - 1.5 * IQR
    upper_bnd = Q3 + 1.5 * IQR

    for i in data:
        if i < lower_bnd or i > upper_bnd:
            outliers.append(i)

    return list(set(outliers)), lower_bnd, upper_bnd
```

```python
outliers,lower_bnd,upper_bnd = detect_outliers(df['COURSE 1 MARKS'])
```

```python
df['cleaned_course_1_marks'] = df['COURSE 1 MARKS'].apply(lambda x : x if lower_bnd<=x <= upper_bnd else np.nan)
df.head()
```
